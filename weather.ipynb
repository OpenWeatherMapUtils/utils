# Weather (from openweathermap.org) @ IMAG (52 weeks since June 2021)

```python
#
# Chart weather from CSV files
# 
# See https://openweathermap.org/api/

# TODO explore graphs for dataviz https://plotly.com/python/

# Parameter(s)
poi='IMAG'

# Import the necessaries libraries
import plotly.offline as pyo
import plotly.graph_objs as go
# Set notebook mode to work in offline
pyo.init_notebook_mode()

import pandas as pd

import glob
import os

#
# merge all CSV in one dataframe
#
  
# A list of all joined files is returned
joined_files = os.path.join("carbon*.json.csv")
joined_files = os.path.join("weather", "weather-"+poi+"-*.csv")
joined_list = glob.glob(joined_files)
# print(joined_list)

# Finally, the files are joined
weather = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
weather = weather.sort_values(by=['dt'])

dtfirst = weather.iloc[0]['dt']
weather['dt'] = ((weather['dt'] - dtfirst)/ 1000)

#print(weather)
#print(weather.columns)

# TODO divide 'dt' by 1000 and substract first 'dt'

# Create traces

# TODO add options https://plotly.com/python-api-reference/generated/plotly.graph_objects.Scatter.html

# temp', 'feels_like', 'pressure', 'humidity', 'temp_min',
#       'temp_max', 'wind_speed', 'wind_deg', 'wind_gust', 'cloudiness',
    
    
temperature = go.Scatter(
    x=weather.dt,
    y=weather.temp,
    name='Temperature (°C)'
)
temp_max = go.Scatter(
    x=weather.dt,
    y=weather.temp_max,
    name='Temp. Max (°C)'
)
temp_min = go.Scatter(
    x=weather.dt,
    y=weather.temp_min,
    name='Temp. Min (°C)'
)


feels_like = go.Scatter(
    x=weather.dt,
    y=weather.feels_like,
    name='Feels like (°C)'
)

humidity = go.Scatter(
    x=weather.dt,
    y=weather.humidity,
    name='Humidity (%)'
)

wind_speed = go.Scatter(
    x=weather.dt,
    y=weather.wind_speed,
    name='Wind speed (m/s)'
)

wind_deg = go.Scatter(
    x=weather.dt,
    y=weather.wind_deg,
    name='Wind orientation (°)'
)

cloudiness = go.Scatter(
    x=weather.dt,
    y=weather.cloudiness,
    name='Cloudiness (%)'
)



# TODO Plot a map here with lat,lon
#lat = weather.iloc[0]['lat']
#lon = weather.iloc[0]['lon']

# Plot it and save as basic-line.html
pyo.iplot([temperature,feels_like,temp_max,temp_min], filename = 'temperature')
pyo.iplot([humidity], filename = 'humidity')
pyo.iplot([wind_speed], filename = 'wind_speed')
pyo.iplot([wind_deg], filename = 'wind_deg')
pyo.iplot([cloudiness], filename = 'cloudiness')

```
